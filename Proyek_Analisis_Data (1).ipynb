## Visualization & Explanatory Analysis

### Pertanyaan 1:
Berapa banyak uang yang dikeluarkan oleh pelanggan setiap tahun?

```python
all_data['order_approved_at'] = pd.to_datetime(all_data['order_approved_at'])
years_spend_df = all_data.resample(rule='Y', on='order_approved_at').agg({
    "payment_value": "sum"
})
years_spend_df.index = years_spend_df.index.strftime('%Y')
years_spend_df = years_spend_df.reset_index()
years_spend_df.rename(columns={
    "payment_value": "total_spend"
}, inplace=True)
years_spend_df.head()
```

```python
years_spend_df = all_data.set_index('order_approved_at').resample(rule='Y').agg({
    "payment_value": "sum"
})

plt.figure(figsize=(10, 5))
colors = ["#72BCD4", "#D3D3D3", "#068DA9"]

plt.bar(years_spend_df.index.year, years_spend_df['payment_value'], color=colors)
plt.title('Total Pengeluaran Pelanggan per tahun', fontsize=15)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Total Pengeluaran', fontsize=12)
plt.xticks(years_spend_df.index.year)
plt.grid(axis='y')
plt.show()
```

### Pertanyaan 2:
Negara mana saja yang memiliki pelanggan terbanyak?

```python
bystate_df = all_data.groupby(by="customer_state").customer_id.nunique().reset_index().head(5)
bystate_df.rename(columns={
    "customer_id": "customer_count"
}, inplace=True)
bystate_df
plt.figure(figsize=(10, 5))
palettes = ["#068DA9" if score == bystate_df.index[0] else "#D3D3D3" for score in bystate_df.index]

sns.barplot(
    x="customer_count",
    y="customer_state",
    data=bystate_df.sort_values(by="customer_count", ascending=False),
    hue="customer_state",
    palette = palettes
)
plt.title("Pelanggan Terbanyak berdasarkan Negara", loc="center", fontsize=15)
plt.ylabel("Asal Pelanggan")
plt.xlabel("Jumlah Pelanggan")
plt.tick_params(axis='y', labelsize=12)
plt.show()
```

## Pertanyaan 3:
Berapa tingkat kepuasan terhadap customer?

```python
review_scores = all_data['review_score'].value_counts().sort_values(ascending=False)

most_common_score = review_scores.idxmax()

sns.set(style="darkgrid")

plt.figure(figsize=(10, 5))

palette = ["#068DA9" if i == review_scores.index[4] else "#D3D3D3" for i in review_scores.index]

sns.barplot(x=review_scores.index,
            y=review_scores.values,
            order=review_scores.index,
            hue=review_scores.values,
            palette=palette
            )

plt.title("Tingkat Kepuasan Pelanggan", fontsize=15)
plt.legend(title='Jumlah Pelanggan Yang Sudah Review', fontsize=10)
plt.xlabel("Rating")
plt.ylabel("Jumlah Pelanggan")
plt.xticks(fontsize=12)
plt.show()
```
